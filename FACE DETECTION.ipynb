```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

# Create Classifiers

```python
face_classifier=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

# open image

```python
image=cv2.imread('rahul-gandhi_1526743864.jpeg')
```

```python




fix_image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_image)


```

# Detect Faces

```python
faces = face_classifier.detectMultiScale(fix_image , 1.4, 5)
```

# No Faces detected

```python
if faces is ():
    print('No faces detected')
    
```

# Now the magic

```python
def detect_face(fix_image):
    face_rects=face_classifier.detectMultiScale(fix_image)
    
    for(x,y,w,h) in face_rects:
        cv2.rectangle(fix_image,
                     (x,y),
                     (x+w , y+h),
                     (255,0,0),
                     5) 
    return fix_image
```

```python
result=detect_face(fix_image)
plt.imshow(result)
```
