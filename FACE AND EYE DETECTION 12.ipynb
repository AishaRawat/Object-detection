```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
face_classifier=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_classifier=cv2.CascadeClassifier('haarcascade_eye.xml')
```

```python
image=cv2.imread('rahul-gandhi_1526743864.jpeg')
fix_image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
faces = face_classifier.detectMultiScale(fix_image , 1.4, 5)
```

```python
if faces is ():
    print('No faces detected')
```

```python
def detect_face_eyes(fix_image):
    face_rects=face_classifier.detectMultiScale(fix_image)
    
    for(x,y,w,h) in face_rects:
        cv2.rectangle(fix_image,
                     (x,y),
                     (x+w , y+h),
                     (255,0,0),10)
        
        eyes_rects=eye_classifier.detectMultiScale(fix_image)
        for(ix,iy,iw,ih) in eyes_rects:
            cv2.rectangle(fix_image,
                     (ix,iy),
                     (ix+iw , iy+ih),
                     (255,255,255),
                     10) 
    return fix_image
```

```python
result=detect_face_eyes(fix_image)
plt.imshow(result)
```
